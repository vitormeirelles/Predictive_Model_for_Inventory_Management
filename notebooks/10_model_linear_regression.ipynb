```python
import pandas as pd
```

**Load the tables**

```python
df = spark.sql("SELECT * FROM ML_Area_.features.features_table")
display(df)
```

```python
features_pd = df.toPandas()
```

**Filter out store 95 and product type  H00**

```python
features_pd = features_pd[features_pd['store_id'] == 'S0080']
```

```python
features_pd = features_pd[features_pd['hierarchy1_id'] == 'H00']
```

```python
features_pd = features_pd[features_pd['city_name'] == 'Istanbul']
```

```python
features_pd.head(100)
```

```python
print("Columns in features_pd:")
for col in features_pd.columns:
    print(col)
```

**Columns to drop**

```python
columns_to_drop = [
    "hierarchy2_id",
    "hierarchy3_id",
    "hierarchy4_id",
    "hierarchy5_id",
    "storetype_id",
    "cluster_id",
    "num_obs"
]

# Drop the columns
features_pd = features_pd.drop(columns=columns_to_drop)

# Confirm columns were removed
print("Remaining columns:\n", features_pd.columns)
```

**One-hot encoding**

```python
cols_to_encode = [
    "promo_bin_1_rate",
    "promo_bin_2_rate",
]

# One-hot encode the specified columns
features_pd = pd.get_dummies(features_pd, columns=cols_to_encode, prefix=cols_to_encode)
```

```python
cols_to_encode = [
    "year"
]

# One-hot encode the specified columns
features_pd = pd.get_dummies(features_pd, columns=cols_to_encode, prefix=cols_to_encode)
```

```python
features_pd.head()
```

**TURN NaN into 0s**

```python
cols_to_fill = ['lag_sales_1w', 'lag_sales_2w', 'lag_sales_3w']
features_pd[cols_to_fill] = features_pd[cols_to_fill].fillna(0)
```

**TRAIN TEST SPLIT**

```python
# Ensure the data is sorted
features_pd = features_pd.sort_values("week_start").reset_index(drop=True)

# Find a split date that creates an 80/20 split
split_index = int(len(features_pd) * 0.8)
split_date = features_pd.loc[split_index, "week_start"]

# Split using the split_date (exclusive)
train = features_pd[features_pd["week_start"] < split_date]
test = features_pd[features_pd["week_start"] >= split_date]

# Check the ranges
print("Train weeks:", train["week_start"].min(), "to", train["week_start"].max())
print("Test weeks:", test["week_start"].min(), "to", test["week_start"].max())
```

**Regressao Varias variaveis**

```python
# Define features (X) and target (y)
features = [col for col in features_pd.columns if col not in ['store_id', 'weekly_sales', 'product_id', 'hierarchy1_id', 'city_name', 'week_start']]
target = 'weekly_sales' # Assuming 'sales' is the intended target, but it's missing.

print("Features to be used:", features)
print("Target variable:", target)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train[features], train[target])
```

```python
predictions = model.predict(test[features])
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Predict on the training set
train_predictions = model.predict(train[features])

# Evaluation metrics on the training set
r2_train = r2_score(train[target], train_predictions)
rmse_train = np.sqrt(mean_squared_error(train[target], train_predictions))
mae_train = mean_absolute_error(train[target], train_predictions)

# Print results
print("Train RÂ²:", r2_train)
print("Train RMSE:", rmse_train)
print("Train MAE:", mae_train)
```

**TIME-SERIES CROSS VALIDATION**

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Define features and target
X = features_pd[features]
y = features_pd[target]

# Time series cross-validator
tscv = TimeSeriesSplit(n_splits=5)

# Initialize model
model = LinearRegression()

# Store scores
r2_scores = []
rmse_scores = []
mae_scores = []

# Time series cross-validation
for train_index, test_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Impute missing values (mean of training set)
    for feature in features:
        if X_train[feature].isna().any():
            mean_value = X_train[feature].mean()
            X_train.loc[:, feature] = X_train.loc[:, feature].fillna(mean_value)
            X_test.loc[:, feature] = X_test.loc[:, feature].fillna(mean_value)

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)

    r2_scores.append(r2)
    rmse_scores.append(rmse)
    mae_scores.append(mae)

    print(f"Fold RÂ²: {r2:.4f}, Fold RMSE: {rmse:.4f}, Fold MAE: {mae:.4f}")

# Average metrics
print("\nAverage RÂ²:", np.mean(r2_scores))
print("Average RMSE:", np.mean(rmse_scores))
print("Average MAE:", np.mean(mae_scores))


```
